### 실험 결과 비교

```python
# 실험 결과 비교
results = []
for f in sorted(glob.glob('../saved_models/*_result.json')):
    results.append(json.load(open(f, encoding='utf-8')))

df = pd.DataFrame(results)[['name', 'mean_roc_auc', 'mean_pr_auc']]
df = df.sort_values('mean_roc_auc', ascending=False).reset_index(drop=True)
print(df.to_string(index=False))
```
