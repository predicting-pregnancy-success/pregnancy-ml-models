### 환경 설정

```python
import numpy as np
import pandas as pd
```

### v1 — 튜닝 앙상블 제출
Best Weights: LightGBM 0.3 / XGBoost 0.4 / CatBoost 0.3  
리더보드 점수: 0.7410583256

```python
lgbm_test = np.load('../saved_models/exp08_tuned_lgbm_test.npy')
xgb_test  = np.load('../saved_models/exp09_tuned_xgb_test.npy')
cat_test  = np.load('../saved_models/exp10_tuned_cat_test.npy')

best_weights = (0.3, 0.4, 0.3)

submission = pd.read_csv('../data/raw/sample_submission.csv')
submission['probability'] = best_weights[0]*lgbm_test + best_weights[1]*xgb_test + best_weights[2]*cat_test
submission.to_csv('../submissions/submission_v1_tuned_ensemble.csv', index=False)
print(submission.head())
```
